```python
####### LÓGICA PROGRAMA #######
import matplotlib.pyplot
import pandas as pd

def read_dataset(path, encoding_type='ISO-8859-1'):
    return pd.read_csv(path, encoding=encoding_type)

def dataset_visualize_items_specific_columns(dataset, column_name):
    return dataset[column_name].value_counts()

def filter_dataset_by_column(dataset, column_name, item_column):
    return dataset.loc[dataset[column_name] == item_column]

def obtain_mean_group_of_numerical_column_by_day(dataset, column_name):
    return dataset.groupby(column_name).mean()

def obtain_mean_group_of_numerical_column_by_month(dataset, column_name, result):
    dataset[column_name] = pd.to_datetime(dataset[column_name])
    return dataset.groupby(dataset[column_name].dt .month) [result]. mean()

def obtain_mean_group_of_numerical_column_by_year(dataset, column_name):
    return dataset.groupby(column_name).mean()

def return_year_dataset(dataset, column_name):
    dataset[column_name] = pd.to_datetime(dataset[column_name])
    return int(dataset[column_name].dt .year.head(1))
```
